Combining all temperature columns into one average temperature column. Dropping all other temperature columns

```python
temp_columns = ['Seville_temp_max','Madrid_pressure', 'Valencia_temp_max', 'Valencia_temp', 'Seville_temp','Valencia_temp_min', 'Barcelona_temp_max', 'Madrid_temp_max','Barcelona_temp', 'Bilbao_temp_min', 'Bilbao_temp','Barcelona_temp_min', 'Bilbao_temp_max', 'Seville_temp_min','Madrid_temp', 'Madrid_temp_min']
df_mean = df[temp_columns]
temp_means = df_mean.mean(axis=1)
df.insert(len(df.columns), 'cities_average_temp', temp_means, allow_duplicates=False)
df = df.drop(temp_columns, axis=1)
df.head()
```

The columns Madrid_humidity and Seville_humidity are also combined by calculating the average for each row. The original columns are dropped

```python
humid_columns = ['Madrid_humidity', 'Seville_humidity']
humid_mean_df = df[humid_columns]
humid_means = humid_mean_df.mean(axis=1)
df.insert(2, 'Madrid_Seville_humidity', humid_means, allow_duplicates=False)
df = df.drop(humid_columns, axis=1)
df.head()
```

Here we are re-engineering the time column by creating 5 new columns: year, month, day, hour and day_of_week. Each of these columns will contain numerical data and the original time column will be dropped. The reason for adding day of the week as a feature is that it may be a factor when predicting load_shortfall_3h. In South Africa during loadshedding, the weekends are often when load-shedding is reduced or eliminated as power consumption is usually reduced.

```python
def extract_year(datetime):
    temp = datetime.split(" ")
    date = None
    if '/' in temp[0]:
        date = temp[0].split('/')
    else:
        date = temp[0].split('-')
    return int(date[0])

def extract_month(datetime):
    temp = datetime.split(" ")
    date = None
    if '/' in temp[0]:
        date = temp[0].split('/')
    else:
        date = temp[0].split('-')
    return int(date[1])

def extract_day(datetime):
    temp = datetime.split(" ")
    date = None
    if '/' in temp[0]:
        date = temp[0].split('/')
    else:
        date = temp[0].split('-')
    return int(date[2])

def extract_hour(datetime):
    temp = datetime.split(" ")
    date = None
    if '/' in temp[0]:
        date = temp[0].split('/')
    else:
        date = temp[0].split('-')
    time = temp[1].split(':')
    return int(time[0])

def extract_day_of_week(info):
    temp = info.split(" ")
    date = None
    if '/' in temp[0]:
        date = temp[0].split('/')
    else:
        date = temp[0].split('-')
    time = temp[1].split(':')
    ans = datetime.date(int(date[0]), int(date[1]), int(date[2]))
    weekday = ans.isocalendar()[2]
    return weekday

df['Year'] = df['time'].apply(extract_year)
df['Month'] = df['time'].apply(extract_month)
df['Day'] = df['time'].apply(extract_day)
df['Hour'] = df['time'].apply(extract_hour)
df['Day_of_Week'] = df['time'].apply(extract_day_of_week)
```

Dropping all columns with text data

```python
df = df.drop('time', axis=1)
df = df.drop('Seville_pressure', axis=1)
df = df.drop('Valencia_wind_deg', axis=1)
df = df.drop('Valencia_pressure', axis=1)
```

The columns in the dataframe are now not in an ideal order, we are going to move the new time columns to the left hand side and the load_shortfall_3h column will be moved into a separate dataset. Essentially, we are making targets and features datasets

```python
df = df[['Year', 'Month', 'Day', 'Hour', 'Day_of_Week', 'Madrid_wind_speed', 'Madrid_Seville_humidity', 
'Bilbao_rain_1h', 'Valencia_wind_speed', 'Bilbao_clouds_all', 'Bilbao_wind_speed',
 'Seville_clouds_all', 'Bilbao_wind_deg', 'Barcelona_wind_speed',
 'Barcelona_wind_deg', 'Madrid_clouds_all', 'Seville_wind_speed',
 'Barcelona_rain_1h', 'Seville_rain_1h', 'Bilbao_snow_3h',
 'Barcelona_pressure', 'Seville_rain_3h', 'Madrid_rain_1h',
 'Barcelona_rain_3h', 'Valencia_snow_3h', 'Madrid_weather_id',
 'Barcelona_weather_id', 'Bilbao_pressure', 'Seville_weather_id',
 'Bilbao_weather_id', 'Valencia_humidity', 'cities_average_temp', 'load_shortfall_3h']]

X = df[['Year', 'Month', 'Day', 'Hour', 'Day_of_Week', 'Madrid_wind_speed', 'Madrid_Seville_humidity', 
'Bilbao_rain_1h', 'Valencia_wind_speed', 'Bilbao_clouds_all', 'Bilbao_wind_speed',
 'Seville_clouds_all', 'Bilbao_wind_deg', 'Barcelona_wind_speed',
 'Barcelona_wind_deg', 'Madrid_clouds_all', 'Seville_wind_speed',
 'Barcelona_rain_1h', 'Seville_rain_1h', 'Bilbao_snow_3h',
 'Barcelona_pressure', 'Seville_rain_3h', 'Madrid_rain_1h',
 'Barcelona_rain_3h', 'Valencia_snow_3h', 'Madrid_weather_id',
 'Barcelona_weather_id', 'Bilbao_pressure', 'Seville_weather_id',
 'Bilbao_weather_id', 'Valencia_humidity', 'cities_average_temp']]
y = df[['load_shortfall_3h']]
```

## Model 3:Random Forest with K-fold Cross-validation
<a class="anchor" id="1.1"></a>
<a href=#cont>Back to Table of Contents</a>

Targets and features datasets must be converted to np arrays to work with Kfold method

```python
x_numpy = X.to_numpy()
y_numpy = y.to_numpy()
```

SkLearn's KFold class is used to generate the datasets needed for model training and cross-validation

```python
kf = KFold(n_splits=KFOLD, shuffle=True, random_state=PARAMETER_CONSTANT)
X_train_RF, X_test_RF, y_train_RF, y_test_RF = [],[],[],[]
for train_index, test_index in kf.split(x_numpy):
    X_train_RF += [x_numpy[train_index]]
    X_test_RF += [x_numpy[test_index]]
    y_train_RF += [y_numpy[train_index]]
    y_test_RF += [y_numpy[test_index]]
```

Based on the value of the KFOLD constant, multiple train-test datasets have been created above. Below, SKLearn's RandomForestRegressor class is used to create and train KFOLD Random Forest models. WARNING: This cell will take 1-2 minutes to run due to the computational cost required to train multiple Random Forests

```python
model_list = []

for i in range(KFOLD):
    RF = RandomForestRegressor(n_estimators=100, max_depth=8, random_state= PARAMETER_CONSTANT) #n_estimators=100, max_depth=8 gives lowest rmse on kaggle
    XTr = X_train_RF[i]
    YTr = y_train_RF[i]
    XTe = X_test_RF[i]
    YTe = y_test_RF[i]
    #Train model
    RF.fit(XTr,np.ravel(YTr))
    model_list.append(RF)
```

Here we are using our models created above to predict y-values for the test set. MSE, RMSE and R-squared values for each model's predictions are determined. The model with the lowest RMSE is chosen as the Random Forest best model for the following step we evaluate model performance

```python
model_mse = []
model_rmse = []
model_R2 = []
for i in range(KFOLD):
    model = model_list[i]
    #print("Model " + str(i+1) + ":")
    y_pred_val = model.predict(X_test_RF[i])
    mse = metrics.mean_squared_error(y_test_RF[i], y_pred_val)
    model_mse += [mse]
    #print('MSE:', mse)
    rmse = np.sqrt(metrics.mean_squared_error(y_test_RF[i], y_pred_val))
    model_rmse += [rmse]
    #print('RMSE:', rmse)
    R2 = metrics.r2_score(y_test_RF[i], y_pred_val)
    model_R2 += [R2]
    #print('R_squared:', R2)

min_rmse = min(model_rmse)
model_idx = model_rmse.index(min_rmse)
RF_model = model_list[model_idx]
RF_mse = model_R2[model_idx]
RF_rmse = min_rmse
RF_R2 = model_R2[model_idx]
```

```python
val_dict = {'model mse':model_mse, 'model rmse': model_rmse, 'model R2': model_R2}
val_df = pd.DataFrame(val_dict)
```

As shown in the figure below, the prediction accuracy of random forests varies considerably for any given PARAMETER CONSTANT value. RMSE values for the 10 models produced varies between 3000 and 3500 when evaluating using the test datasets. It is therefore a good idea to train multiple Random Forest models and pick one or more that perform the best on some holdout dataset.

Shown in the figure below, the random forest model produces by far, the highest R-squared value during validation with each model's corresponding test dataset. R-squared or the coefficient of determination is a measure of the proportion of variance in the response variable (load_shortfall_3h) that can be collectively explained by the predictor variables (X-train, X_test datasets). A higher R-squared value indicates that a model accounts for more of the variance in an input dataset. However a low R-squared value for a model does not necessarily mean a model is flawed, datasets with large amounts of noisy data (high variance) will result in a lower R-squared value. Given the noise in this dataset, it is likely that both model 1 and 2, being linear models struggled to fit to the data sufficiently while model 3, being a non-linear decision tree was able to fit to the data significantly better

Our Random forest model was produced in two steps. The first step involves a simplified implementation of a technique known as K-fold cross-validation. Essentially, we split our df_train.csv dataset into training and testing subsets 10 times (k = 10, so 10-fold cross-validation). Each test subset was approximately 10% of all datapoints in the dataset and contained datapoints that weren't in the test subsets of any of the other 10 folds. Essentially, each datapoint only appeared in one test subset. Refer the graphic below taken from SKLearn's User Guide for a visualization of the K-fold technique, note how every test subset in each fold is comprised of different observations.
![SKLearn Kfold image](https://scikit-learn.org/stable/_images/sphx_glr_plot_cv_indices_006.png)

Normally, the K-fold technique is used to validate a single model, essentially allowing every datapoint to be part of the training and testing stages. For the model trained in this notebook however, the K-fold technique was used to essentially create 10 unique train/test splits of the dataset, which were then used to train and test 10 different Random Forest models.

To understand how Random Forests work, its best to start with the basics, and a basic version of a Random Forest is a Decision Tree. Decision Trees are a type of non-linear prediction model that works by partitioning data into subsets and learning simlple decision rules inferred from features in the dataset. In the image below, taken from SKLearn's User Guide, a decision tree is trained to generate a sine curve using a logic tree of simple if, then, else decisions based on the value of a single predictor variable on the x-axis
![SKLearn Kfold image](https://scikit-learn.org/stable/_images/sphx_glr_plot_tree_regression_001.png)
Random Forests can be thought of as a forest of Decision Trees - hence their name. They are a type of ensemble model that makes use of tens or even hundreds of decision trees which, for a given set of observations will each make a prediction. These predictions in the ensemble of a random forest are then averaged to produce the final output of the model. By using the power of consensus, Random Forest models can avoid one of the main disadvantages of Decision Trees: they tend to easily overfit data and then perform significantly worse on unseen data as they haven't learnt the general trend in the data. However they are also significantly more computationally expensive to train, as training one means you are effectively training tens or hundreds of decision trees
